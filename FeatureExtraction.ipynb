```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('finaldata.pkl','rb'))
df
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
df['city']=np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df.isnull().sum()
```

```python
df.drop('venue',axis=1,inplace=True)
```

```python
df['city'].value_counts()[df['city'].value_counts() > 1500].sum()
```

```python
eligible = df['city'].value_counts()[df['city'].value_counts() > 1500].index.to_list()
```

```python
df = df[df['city'].isin(eligible)]
df
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype(int)
```

```python
df['balls_left'] = 300 - df['balls_bowled']
```

```python
df
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x == '0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype(int)
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
df.columns
```

```python
df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
df = df[['match_id','batting_team', 'bowling_team','city','balls_left', 'wickets_left', 'crr', 'last_five','runs_x' ]]
df.rename(columns={'runs_x':'final_score'})
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
df = df.sample(df.shape[0])
```

```python
X = df[['batting_team', 'bowling_team', 'city', 'balls_left',
       'wickets_left', 'crr', 'last_five']]
Y = df[['runs_x']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
!pip install xgboost
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,Y_train)
Y_pred = pipe.predict(X_test)
print(r2_score(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
